## Import Libraries

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

## Load Dataset

```python
# Load the Titanic dataset
df_titanic = sns.load_dataset('titanic')

# Convert to DataFrame
df = pd.DataFrame(df_titanic)
```

```python
df.head()
```

### Cherry-Picking Data (Bar Plot Only for 1st Class Passengers):
    By only focusing on 1st-class passengers, we could create the impression that survival rates were generally much higher.

```python
# Filter the data for 1st-class passengers
first_class = df[df['pclass'] == 1]

# Create a bar plot for survival rate by gender for 1st-class passengers
sns.barplot(x='sex', y='survived', data=first_class)
plt.title('Survival Rate by Gender (1st Class Only)')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
```

### Manipulating the y-scale

```python
sns.barplot(x='sex', y='survived', data=first_class)
plt.ylim(0.4, 1)  # Manipulating the y-axis to exaggerate the differences
plt.title('Misleading Survival Rate by Gender (1st Class Only)')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
```

### Accurate Visualization: Survival Rate by Gender

```python
# Create a bar plot for survival rate by gender
sns.barplot(x='sex', y='survived', data=df)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
```

### How many people were on the titanic?

```python
titanic_passenger_count = df.shape[0]
print(titanic_passenger_count)
```

### Survival Rate for people on 1st-class

```python
# Filter the data for 1st-class passengers
first_class_data = df_titanic[df_titanic['pclass'] == 1]

# Calculate the survival rate for 1st-class passengers
survival_rate_first_class = first_class_data['survived'].mean()

# Display the survival rate
print(survival_rate_first_class)
```

```python
# Create a bar graph for the survival rate of 1st-class passengers
sns.barplot(x=["1st Class"], y=[survival_rate_first_class], palette="Blues")

# Adding title and labels
plt.title('Survival Rate for 1st-Class Passengers')
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')

# Display the plot
plt.show()
```

### Survival Rate for All

```python
# Calculate the survival rate for all passengers
survival_rate_all = df_titanic['survived'].mean()

# Display the survival rate for everyone
print(survival_rate_all)
```

```python
# Create a bar graph for the survival rate of all passengers
sns.barplot(x=["All Passengers"], y=[survival_rate_all], palette="Blues")

# Adding title and labels
plt.title('Survival Rate for All Passengers')
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Group')

# Display the plot
plt.show()
```
