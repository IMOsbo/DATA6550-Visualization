### Cherry-Picking Data (Bar Plot Only for 1st Class Passengers):
    By only focusing on 1st-class passengers, we could create the impression that survival rates were generally much higher.

```python
# Filter the data for 1st-class passengers
first_class_data = df[df['pclass'] == 1]

# Create a bar plot for survival rate by gender for 1st-class passengers
sns.barplot(x='sex', y='survived', data=first_class_data, palette='coolwarm')

# Exaggerating the y-axis range to make differences look bigger
plt.ylim(0.4, 1)  # Limiting the range to exaggerate survival differences

# Adding title and labels
plt.title('Survival Rate for 1st-Class Passengers by Gender (Misleading)', fontsize=16, weight='bold')
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Survival Rate (%)', fontsize=12)

# Add annotations to emphasize the difference
for i, v in enumerate(first_class_data.groupby('sex')['survived'].mean()):
    plt.text(i, v + 0.05, f'{v*100:.1f}%', ha='center', fontsize=14, fontweight='bold')

# Display the plot
plt.show()
```

### Adding fare in the y-axis

```python
# Create age groups to add complexity
bins = [0, 18, 40, 60, 100]
labels = ['<18', '18-40', '40-60', '60+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

# Group the data by multiple factors: passenger class, gender, and age group
grouped_data = df.groupby(['pclass', 'sex', 'age_group']).agg({'survived': 'mean', 'fare': 'mean'}).reset_index()

# Create the plot
fig, ax1 = plt.subplots(figsize=(12, 8))

# Create a bar plot for survival rate (left y-axis)
sns.barplot(x='age_group', y='survived', hue='sex', data=grouped_data, ci=None, palette='Set2', 
            hue_order=['male', 'female'], ax=ax1)

# Add a misleading y-axis range for survival rate starting at 0.2
ax1.set_ylim(0.2, 1.0)  # Set the survival rate y-axis from 0.2 to 1
ax1.set_ylabel('Survival Rate', fontsize=12)

# Set custom y-axis ticks for survival rate at 0.2, 0.4, 0.6, ..., 1.0
ax1.set_yticks([i * 0.2 for i in range(1, 6)])

# Add a second y-axis on the right for the fare
ax2 = ax1.twinx()

# Plot the fare on the second y-axis
sns.lineplot(x='age_group', y='fare', data=grouped_data, ax=ax2, color='green', marker='o', linewidth=2)

# Set the y-axis for fare, scaling it to fit visually with the survival rate
ax2.set_ylabel('Fare', fontsize=12)
ax2.set_ylim(0, grouped_data['fare'].max() + 50)  # Adjust the range for the fare y-axis if needed

# Adding title and formatting
plt.title('Survival Rates and Fare Across Multiple Factors', fontsize=16, weight='bold')
ax1.set_xlabel('Age Group', fontsize=12)

# Add a legend to the right side for the fare plot
ax2.legend(['Fare'], loc='upper right')

# Display the plot
plt.show()
```

### Add sex on the x-axis. This will make it even more confusing whe combining age groupo and sex on the x-axis. Also class is included in the chart.

### Survival rate of each class as a diameter

```python
# Calculate survival rate for each class
class_survival_rate = df.groupby('pclass')['survived'].mean()

# Map survival rate to bubble sizes
bubble_sizes = class_survival_rate * 1000  # Scale the size for better visualization

# Create the plot
fig, ax = plt.subplots(figsize=(8, 6))

# Plot bubbles using scatter, where size corresponds to the survival rate
scatter = ax.scatter(class_survival_rate.index, class_survival_rate.values, 
                     s=bubble_sizes, c=bubble_sizes, cmap='coolwarm', alpha=0.6, edgecolors="w", linewidth=2)

# Add labels and title
ax.set_xlabel('Passenger Class', fontsize=12)
ax.set_ylabel('Survival Rate', fontsize=12)
ax.set_title('Survival Rate by Passenger Class (Bubble Chart)', fontsize=16, weight='bold')

# Customize x-axis labels to show class names (1st, 2nd, 3rd)
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(['1st Class', '2nd Class', '3rd Class'])

# Add a colorbar for the bubble colors (representing survival rate)
plt.colorbar(scatter, label='Survival Rate')

# Display the plot
plt.show()
```

### Let's try to make the diameter worse to deceive the human eye.

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Wedge

# Calculate survival rate for each class
class_survival_rate = df.groupby('pclass')['survived'].mean()

# Create figure
fig, ax = plt.subplots(figsize=(10, 6))

# Set the colors for each passenger class
colors = ['lightcoral', 'lightblue', 'lightgreen']  # Different colors for each class
labels = ['1st Class', '2nd Class', '3rd Class']

# Create half-circles (Wedges) for each class, with size corresponding to survival rate
for i, (rate, label) in enumerate(zip(class_survival_rate, labels)):
    radius = rate  # Set radius proportional to the survival rate
    # Create a wedge (half-circle) for each class
    wedge = Wedge(center=(i * 3, 0), r=radius, theta1=0, theta2=180, color=colors[i], label=label, alpha=0.7, edgecolor="black")
    ax.add_patch(wedge)

# Add title and remove axis labels
ax.set_title('Misleading Survival Rates with Semi-Circles (Diameters)', fontsize=16, weight='bold', color='magenta')

# Remove the axes
ax.set_axis_off()

# Set limits to ensure semi-circles fit within the plot area
ax.set_xlim([-2, 10])
ax.set_ylim([-1.5, 1.5])

# Add a legend to confuse the viewer
ax.legend(title="Passenger Classes", loc="best", frameon=True, facecolor='yellow')

# Display the plot
plt.tight_layout()
plt.show()
```
