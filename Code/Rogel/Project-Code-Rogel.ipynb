### Adding fare in the y-axis

```python
# Create age groups to add complexity
bins = [0, 18, 40, 60, 100]
labels = ['<18', '18-40', '40-60', '60+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

# Group the data by multiple factors: passenger class, gender, and age group
grouped_data = df.groupby(['pclass', 'sex', 'age_group']).agg({'survived': 'mean', 'fare': 'mean'}).reset_index()

# Create the plot
fig, ax1 = plt.subplots(figsize=(12, 8))

# Create a bar plot for survival rate (left y-axis)
sns.barplot(x='age_group', y='survived', hue='sex', data=grouped_data, ci=None, palette='Set2', 
            hue_order=['male', 'female'], ax=ax1)

# Add a misleading y-axis range for survival rate starting at 0.2
ax1.set_ylim(0.2, 1.0)  # Set the survival rate y-axis from 0.2 to 1
ax1.set_ylabel('Survival Rate', fontsize=12)

# Set custom y-axis ticks for survival rate at 0.2, 0.4, 0.6, ..., 1.0
ax1.set_yticks([i * 0.2 for i in range(1, 6)])

# Add a second y-axis on the right for the fare
ax2 = ax1.twinx()

# Plot the fare on the second y-axis
sns.lineplot(x='age_group', y='fare', data=grouped_data, ax=ax2, color='green', marker='o', linewidth=2)

# Set the y-axis for fare, scaling it to fit visually with the survival rate
ax2.set_ylabel('Fare', fontsize=12)
ax2.set_ylim(0, grouped_data['fare'].max() + 50)

# Adding title and formatting
plt.title('Survival Rates and Fare Across Multiple Factors', fontsize=16, weight='bold')
ax1.set_xlabel('Age Group', fontsize=12)

# Add a legend to the right side for the fare plot
ax2.legend(['Fare'], loc='upper right')

# Display the plot
plt.show()
```

This chart attempts to combine survival rates and fare information across various age groups and genders, further broken down by passenger class. This distortion could lead viewers to make incorrect conclusions about the importance of the survival rate differences.

**Irrelevant Variable (Fare) Distorting the Analysis:**
Another issue is the inclusion of the line plot for fare, which is plotted alongside the bar plot for survival rate. While survival rates are shown as bars, the fare is presented as a line graph on a secondary y-axis. This combination implies a relationship between survival rates and fare when there is no clear connection in the data. These variables should be analyzed separately, but this dual-axis presentation creates the illusion that survival and fare are closely linked, which can mislead viewers into thinking the fare has a strong impact on survival, when it does not.

**Cluttered Visual Design:**
The chart also suffers from an overly complicated design. The use of vibrant colors for the title, gridlines, and multiple axes detracts from the data itself. The magenta title, cyan gridlines, and green fare line are visually exhausting, drawing the viewers focus away from the key information. Rather than guiding the viewer to the most important trends, these design elements introduce unnecessary complexity. The viewers attention is diverted to the colors and formatting, which makes it harder to interpret the survival rates accurately.

**Conclusion:**
Overall, this visualization is misleading because it distorts the presentation of the data through manipulated axes, unnecessary complexity, and irrelevant variables. A clearer, more ethical approach would involve eliminating the dual y-axis, starting the survival rate axis at 0, and simplifying the chart to focus solely on the relationships between the key variables. ### Survival rate of each class as a diameter

### Let's make the survival rate the size of the diameter making the surface area bigger than it needs to be. This will have huge discrepancies between the surface areas.

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Wedge

# Calculate survival rate for each class
class_survival_rate = df.groupby('pclass')['survived'].mean()

# Create figure
fig, ax = plt.subplots(figsize=(10, 6))

# Set the colors for each passenger class
colors = ['lightcoral', 'lightblue', 'lightgreen']  # Different colors for each class
labels = ['1st Class', '2nd Class', '3rd Class']

# Create half-circles (Wedges) for each class, with size corresponding to survival rate
for i, (rate, label) in enumerate(zip(class_survival_rate, labels)):
    radius = rate  # Set radius proportional to the survival rate
    # Create a wedge (half-circle) for each class
    wedge = Wedge(center=(i * 3, 0), r=radius, theta1=0, theta2=180, color=colors[i], label=label, alpha=0.7, edgecolor="black")
    ax.add_patch(wedge)

# Add title and remove axis labels
ax.set_title('Misleading Survival Rates with Semi-Circles (Diameters)', fontsize=16, weight='bold', color='magenta')

# Remove the axes
ax.set_axis_off()

# Set limits to ensure semi-circles fit within the plot area
ax.set_xlim([-2, 10])
ax.set_ylim([-1.5, 1.5])

# Add a legend to confuse the viewer
ax.legend(title="Passenger Classes", loc="best", frameon=True, facecolor='yellow')

# Display the plot
plt.tight_layout()
plt.show()
```

#### Misleading Visualization: Survival Rates with Stacked Semi-Circles (Diameters)
This chart uses a visually confusing design to represent survival rates across different passenger classes. The visualization relies on stacked semi-circles to represent survival rates, where the diameter of each semi-circle is proportional to the survival rate of each class. While the intent may be to show the variation in survival rates, several design choices make the chart misleading and difficult to interpret correctly.

**Exaggerated Use of Diameters:**
In this chart, the size of the semi-circles (or wedges) is supposed to represent the survival rate of each class. However, by using the diameter to represent the rate, the area of the semi-circles is distorted. The area of a circle (or semi-circle) grows quadratically with the diameter, which means that small differences in diameter will result in larger differences in area. For example, a survival rate of 0.4 may look much larger compared to 0.6, just because the visual representation of the area is not proportional to the data. This could mislead the viewer into thinking the difference between survival rates is much more significant than it actually is.

**Lack of Clear Context and Overcomplicated Design:**
The use of semi-circles stacked vertically makes the chart visually confusing, as its hard to discern whether the stacked shapes are intended to show aggregative data or if they represent independent data points. The stacked nature implies that the classes are related or part of a whole, when, in fact, they are independent categories. The chart could easily mislead viewers into assuming that the survival rates of different classes are part of a combined, stacked value, which distorts the interpretation.

**Distracting Legend and Color Choices:**
The color scheme, while visually appealing, doesn't help clarify the data. The light pastel colors for each class (light coral for 1st class, light blue for 2nd class, and light green for 3rd class) do little to differentiate the survival rates effectively, and the use of a confusing legend, which doesnt show specific values, only adds to the confusion. This approach makes the viewer focus more on the shapes and colors than on the actual data being represented. The confusing legend further misguides the viewer by suggesting that these shapes are more informative than they really are.

**Conclusion:**
This visualization misleads the viewer by using distorted geometric shapes (semi-circles) to represent survival rates, which exaggerates the differences between groups. The stacking of shapes, combined with misleading size representation, obscures the true relationship between the variables. A more accurate visualization would use proportional bar charts or pie charts with clear labels and no unnecessary geometric manipulation. By simplifying the design and using appropriate scaling, the data would be more effectively communicated.

## Survival Rates by Age Group

```python
# Create age groups for better segmentation
bins = [0, 18, 40, 60, 100]
labels = ['<18', '18-40', '40-60', '60+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

# Create a bar plot for survival rate by age group without error bars
plt.figure(figsize=(8, 6))
ax = sns.barplot(x='age_group', y='survived', data=df, palette='muted', ci=None, edgecolor='none')

# Set the title and labels
plt.title('Survival Rate by Age Group', fontsize=14)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)

# Add percentage labels on top of the bars
for p in ax.patches:
    height = p.get_height()
    percentage = round(height * 100, 2)  # Convert the height to percentage
    # Adjust label placement to be above the bar and avoid being cut off
    ax.text(p.get_x() + p.get_width() / 2, height + 0.01, f'{percentage}%', 
            ha='center', fontsize=12, color='black')

# Show the plot
plt.show()
```

**Accurate and Uncomplicated Design:** The chart avoids unnecessary complexity, focusing on the key variable: survival rate across age groups. By using a simple bar plot with clear labels and percentages, the visualization ensures that the viewer can easily understand the trends. Additionally, the absence of unnecessary design elements (such as error bars or multiple axes) makes the chart clean and straightforward. This simplicity helps maintain focus on the key message: the survival rate across different age groups.

**Effective Communication:** This visualization allows viewers to make accurate, data-driven interpretations. It is easy to see that younger passengers (under 18) had a higher survival rate, while passengers over 60 had a lower survival rate. The charts design helps convey these differences clearly and without distortion, ensuring that viewers can trust the insights presented.

**Conclustion:** This bar plot is effective because it is clear, simple, and focused on the essential information.
